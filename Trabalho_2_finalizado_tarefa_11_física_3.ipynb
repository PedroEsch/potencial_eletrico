```python
# Bibliotecas
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Função
def func(V0, m, prec, arredondamento):
  V = np.zeros((m,m))
  Vnovo = np.zeros((m,m))

  for k in range(1, m+1, 1):
    for j in range(1, m+1, 1):
      if k == 1:
        V[j-1][k-1] = V0
      elif k == m or j == 1 or j == m:
        V[j-1][k-1] = 0
      else:
        V[j-1][k-1] = 0.5*V0

  delta = 1
  while delta > prec:
    delta = 0
    for k in range(2, m, 1):
      for j in range(2, m, 1):
        Vnovo[j-1][k-1] = (1/4)*(V[j][k-1] + V[j-2][k-1] + V[j-1][k] + V[j-1][k-2])
        variacao = math.fabs((Vnovo[j-1][k-1]-V[j-1][k-1])/V[j-1][k-1])
        if variacao > delta:
          delta = variacao
          V[j-1][k-1]=Vnovo[j-1][k-1]
  if arredondamento == True: #verificando pedido de arredondamento
    for k in range(1, m+1, 1):
      for j in range(1, m+1, 1):
        V[j-1][k-1] = round(V[j-1][k-1]/prec)*prec

  return V

def funcao(prec, arredondamento):
  V0 = 10 #volts
  V1 = -5  #volts
  V2 = -5  #volts
  m = 31
  m1 = 2
  m2 = 4
  V = np.zeros((m,m))
  Vnovo = np.zeros((m,m))

  for k in range(1, m+1, 1):
    for j in range(1, m+1, 1):
      if j == m:
        V[j-1][k-1] = V0
      elif j == 10+1 and k >= 14+1 and k <= 14+1+m1:
        V[j-1][k-1] = V1
      elif k == 15+1 and j >= 10+1 and j <= 10+1+m2:
        V[j-1][k-1] = V2
      elif k == m or k == 1 or j == 1:
        V[j-1][k-1] = 0
      else:
        V[j-1][k-1] = V0

  delta = 1
  while delta > prec:
    delta = 0
    for k in range(2, m, 1):
      for j in range(2, m, 1):
        if j == 10+1 and k >= 14+1 and k <= 14+1+m1:
          V[j-1][k-1] = V1
        elif k == 15+1 and j >= 10+1 and j <= 10+1+m2:
          V[j-1][k-1] = V2
        else:
          Vnovo[j-1][k-1] = (1/4)*(V[j][k-1] + V[j-2][k-1] + V[j-1][k] + V[j-1][k-2])
          variacao = math.fabs((Vnovo[j-1][k-1]-V[j-1][k-1])/V[j-1][k-1])
          if variacao > delta:
            delta = variacao
            V[j-1][k-1]=Vnovo[j-1][k-1]
  if arredondamento == True: #verificando pedido de arredondamento
    for k in range(1, m+1, 1):
      for j in range(1, m+1, 1):
        V[j-1][k-1] = round(V[j-1][k-1]/prec)*prec

  return V
```

```python
# Resolução do livro
V0 = 1 #volts
mexemplo = 12 #número de linhas e colunas
prec = 0.01 #precisão desejada
arredondamento =  True #arredonda o resultado final

exemplolivro = func(V0, mexemplo, prec, arredondamento) #usufruindo da função previamente definida

print('\n')

ajuste = np.zeros((mexemplo,mexemplo))

for i in range (0, mexemplo, 1):
  for j in range (0, mexemplo, 1):
    ajuste[i][j] = exemplolivro[j][i]

print(ajuste) #ajustando para o formato do exemplo do livro
print('\n')
```

```python
# Resolução da figura 2D do livro
V0 = 1 #volts
mlivro = 22 #número de linhas e colunas
prec = 0.001 #precisão desejada
arredondamento =  False #arredonda o resultado final

livro = func(V0, mlivro, prec, arredondamento) #usufruindo da função previamente definida

Exlivro = np.zeros((mlivro,mlivro))
Eylivro = np.zeros((mlivro,mlivro))

for k in range(2, mlivro, 1):
  for j in range(2, mlivro, 1):
    Exlivro[j-1][k-1] = -(livro[j][k-1]-livro[j-2][k-1])
    Eylivro[j-1][k-1] = -(livro[j-1][k-2]-livro[j-1][k])

# Plotando gráfico
fig = plt.figure(dpi=200)
ax = fig.add_subplot(111)
ax.set_aspect('equal', adjustable='box') #forçando o gráfico a ser quadradro
plt.title('Potencial elétrico \n(Para comparar com o exemplo do livro)')
plt.plot([0, mlivro-1], [0,0], '-k', linewidth = 3)
plt.plot([0, 0], [0, mlivro-1], '-k', linewidth = 3)
plt.plot([mlivro-1,mlivro-1], [0, mlivro-1], '-k', linewidth = 3)
plt.plot([0, mlivro-1], [mlivro-1, mlivro-1], '-r', linewidth = 3)
x = 0
while x < mlivro:
  y = 0
  while y < mlivro:
    plt.quiver(x,y,
               Exlivro[x][y]/math.sqrt((Exlivro[x][y]**2)+(Eylivro[x][y])**2),
               Eylivro[x][y]/math.sqrt((Exlivro[x][y]**2)+(Eylivro[x][y])**2),
               scale = 40, width = 0.002, color='blue', headwidth = 5)
    y = y + 1
  x = x + 1
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Resolução do grupo
prec = 0.01 #precisão desejada
arredondamento =  False #arredonda o resultado final

V = funcao(prec, arredondamento) #usufruindo da função previamente definida

print(V)
```

```python
# Resolução do grupo em 2D
m = 31
m1 = 2
m2 = 4
Ex = np.zeros((m,m))
Ey = np.zeros((m,m))

for k in range(2, m, 1):
  for j in range(2, m, 1):
    Ex[j-1][k-1] = -(V[j][k-1]-V[j-2][k-1])
    Ey[j-1][k-1] = -(V[j-1][k]-V[j-1][k-2])

# Plotando gráfico
fig = plt.figure(dpi=200)
ax = fig.add_subplot(111)
ax.set_aspect('equal', adjustable='box') #forçando o gráfico a ser quadradro
plt.title('Potencial elétrico')
plt.plot([0, 0], [0, m-1], '-k', linewidth = 3)
plt.plot([0, m-1], [30, 30], '-k', linewidth = 3)
plt.plot([0, m-1], [0, 0], '-k', linewidth = 3)
plt.plot([10, 10+m2], [15, 15], '-g', linewidth = 3)
plt.plot([10, 10], [14, 14+m1], color='aqua', linewidth = 3)
plt.plot([30, 30], [0, m-1], '-r', linewidth = 3)
x = 0
while x < m:
  y = 0
  while y < m:
    while x == 10 and y >= 14 and y <= 14+m1:
      y = y + 1
    while y == 15 and x >= 10 and x <= 10+m2:
      y = y + 1
    plt.quiver(x,y,
               0.7*Ex[x][y]/math.sqrt((Ex[x][y]**2)+(Ey[x][y])**2),
               0.7*Ey[x][y]/math.sqrt((Ex[x][y]**2)+(Ey[x][y])**2),
               scale = 40, width = 0.002, color='b', headwidth = 5)
    y = y + 1
  x = x + 1
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
